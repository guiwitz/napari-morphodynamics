# Application
MorphoDynamics is a collection of algorithms to study the dynamics of cell morphology and fluorescence intensity fluctuations over time in 2D single-cell time lapse images.

The analysis of intensity and shape dynamics in single-cell time-lapse images comes with different challenges. It relies not only on an accurate segmentation of the single cell over time, but also on a precise split of the cell in different regions of interest (windows), which have to be tracked over time. napari-morphodynamics gives a user-friendly access to the **morphodynamics** functions, which are developed to perform all the steps necessary to investigate single-cells intensity dynamics:

1. cell segmentation
2. windowing
3. cell- and windows-tracking
4. signal extraction
5. data post-processing
6. data visualization

### napari-convpaint
A precise cell segmentation is not only the first step but also the foundation of this analysis, therefore napari-morphodynamics implements state-of-the art segmentation algorithms, such as cellpose, but also an in-house developed deep learning based pixel classifier called [convpaint](https://github.com/hinderling/napari_pixel_classifier). The functionality of convpaint can be used indipendently of morphodynamics, and therefore are wrapped in a dedicated napari plugin, please refer to [**napari-convpaint**](https://github.com/guiwitz/napari-convpaint).

### napari-roidynamics
As some of the processes in this plugin are computationally heavy, we have developed a spin-off plugin called napari-roidynamics, which allows to generate fixed geometries instead of precise segmentation and analyses intensity fluctuations in these regions of interest. Please refer to [**napari-roidynamics**](https://stojiljkovicvetana.github.io/napari-roidynamics/README.html) if you wish to apply intensity dynamics measurements in fixed geometries.

# Examples

napari-moprhodynamics is used to extract meaningful information from time-lapse images of single cells.

Danuser Lab and Pertz Lab both rely on advanced microscopy imaging and image analysis techniques to quantify intracellular dynamics. They study very defined molecular signaling dynamics in order to better understand the underlying mechanisms of cell morphogenesis and fate decision. Over the years they have developed many tools to advance this specific field, such as molecular probes to quantify intracellular signaling, but also advanced imaging techniques involving optogenetic stimulation.

A milestone in their literature, which exhaustively illustrate the need of a tools such as the MorphoDynamics package is represented by the article published in Nature in 2009 by Machacek et. al, entitled [**"Coordination of Rho GTPase activities during cell protrusion"**](https://www.nature.com/articles/nature08242).

This partial reproduction of Figure 2 of the above cited paper, demonstrate how windowing allow the precise quantification of cell movement and correlate it with the expression of a biosensor for the Rac1 protein. Later more in depth analysis can be conducted, such as cross-correlation coefficient of the two variables or correlation over time lag.
If this analysis is conducted for different proteins it can be inferred which proteins are responsible for initializing for example the movement in the leading edge of the cell.
